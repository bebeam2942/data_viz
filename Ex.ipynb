```
def Grade(point) :
      if point<0 :
         print(f' error')
      elif 0 <= point < 50 :
         print(f'  F ')
         return 'F'
      elif 50 <= point < 55 :
         print(f'  D ')
         return 'D'
      elif 55 <= point < 60 :
         print(f'  D+ ')
         return 'D+'
      elif 60 <= point < 65 :
         print(f'  C ')
         return 'C'
      elif 65 <= point < 70 :
         print(f'  C+ ')
         return 'C+'
      elif 70 <= point < 75 :
         print(f'  B ')
         return 'B'
      elif 75 <= point < 80 :
         print(f'  B+ ')
         return 'B+'
      elif 80 <= point <101 :
         print(f'  A ')
         return 'A'
      elif  point > 100 :
         print(f' คะแนนเท่ากับ {point}  error ')
```

```
list_M = []     
for i in B :
    P = Grade(i) 
    list_M.append(P)
list_M
```

```
df_house['Grade'] = pd.DataFrame(list_M) #การเพิ่มcolumnsใหม่
df_house
```
